# Scraping File SMA Sutomo 1 yang lulus jalur SNMPTN

```python
from urllib.request import urlopen
url = 'http://sutomo-mdn.sch.id/sutomo1/'
page = urlopen(url).read()
print(page)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page, "html5lib")
print(soup.prettify)
```

```python
soup.title.text
```

### Find All link

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

### Find Content type

```python
all_latestnews =soup.find_all('li', class_='latestnewscolor-white')
print(all_latestnews)
```

```python
for news in all_latestnews :
    print(news.a.string)
```

### Get Content Link

```python
import re
for news in all_latestnews :
    if(re.search("SNMPTN TP. 2017/2018",news.a.string)) :
        snmptn_link = news.a.get("href")
        
print(url+snmptn_link)
```

```python
page2 = urlopen(url+snmptn_link).read()
soup2 = BeautifulSoup(page2, "html5lib")
print(soup2.prettify)
```

```python
embed_link = soup2.find('embed').get("href")
print(embed_link)
```

### Download PDF

```python
from urllib.request import urlretrieve

def getPDF():    
    urlretrieve(embed_link,"Sutomo 1 SNMPTN 2018.pdf")
    print("PDF Saved")

getPDF()
```

##### PDF is saved in the same place with this file
